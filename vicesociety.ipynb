```python
import requests as req
from bs4 import BeautifulSoup
from csv import writer
```

```python
URL = 'http://vsociethok6sbprvevl4dlwbqrzyhxcxaqpvcqt5belwvsuxaxsutyad.onion/'

TOR_PROXY ={'http':'socks5h://localhost:9050',
            'https':'socks5h://localhost:9050'}

res = req.get(url=URL, proxies = TOR_PROXY)


if res.status_code == 200:
    soup = BeautifulSoup(res.text, 'lxml')
    blocks = soup.find_all("td",valign="top")
    with open(('vicesociety.csv'),'w',encoding = 'utf8', newline='') as f:
        thewriter = writer(f)
        header = ["Title", "Website", "Country", "Description", "Data"]
        thewriter.writerow(header)
    
        for block in blocks[1:]:

            try:
                ## Retrieving bvalues
                bvalues = block.find_all("b")

                ## Retrieving title
                title = bvalues[0]
                print("===========================================================================================")
                print(title.text.strip())
                print("===========================================================================================")
                hrefs = [i['href'] for i in block.find_all('a', href=True)]
                        ## Retrieving website
                web = hrefs[0]

                print(web)

                ## Retrieving country
                country = block.find_all(size="2")[1].text
                print(country)

                ## Description
                desc = bvalues[2]
                print(desc.text.strip())

                ## Visit the block
                URL_2 = hrefs[1]
                res_2 = req.get(url=URL_2, proxies = TOR_PROXY)
                soup_2 = BeautifulSoup(res_2.text, 'html.parser')

                ## Data

                data = soup_2.text
                print(data)
            except:
                print("Error with data in this block")
                continue



            info = [title.text.strip(), web, country, desc.text.strip(), data]
            thewriter.writerow(info)
else:
    print(f'Something went wrong for {URL} with status code : {res.status_code}.')
```
