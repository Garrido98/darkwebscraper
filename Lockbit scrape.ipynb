```python
import requests as req
from bs4 import BeautifulSoup
from csv import writer
```

```python
URL = 'http://lockbitapt6vx57t3eeqjofwgcglmutr3a35nygvokja5uuccip4ykyd.onion/'

TOR_PROXY ={'http':'socks5h://localhost:9050',
            'https':'socks5h://localhost:9050'}

res = req.get(url=URL, proxies = TOR_PROXY)


if res.status_code == 200:
    soup = BeautifulSoup(res.text, 'html.parser')
#     print(soup)
    blocks = soup.find_all("div", class_="post-block")
    
    with open(('lockbit.csv'),'w',encoding = 'utf8', newline='') as f:
        thewriter = writer(f)
        header = ["Title", "Description", "Upload Date", "Updated Date"]
        thewriter.writerow(header)
    
        for block in blocks:
            ## Retrieving link
            link = block.get("onclick")[11:-4]

            ## Retrieving title
            title = block.find("div", class_="post-title")
            print("===========================================================================================")
            print(title.text.strip())
            print("===========================================================================================")

            ## Visit the block
            URL_2 = URL + link
            res_2 = req.get(url=URL_2, proxies = TOR_PROXY)
            soup_2 = BeautifulSoup(res_2.text, 'html.parser')

            ## Description
            desc = soup_2.find("div", class_= "desc")
            print(desc.text.strip())

            ## Upload date
            uploaddate = soup_2.find("div", class_= "uploaded-date-utc")
            print("Upload date:" + uploaddate.text.strip())

            ## Update date
            updatedate = soup_2.find("div", class_= "updated-date-utc")
            print("Updated date: "+updatedate.text.strip())

            info = [title.text.strip(), desc.text.strip(), uploaddate.text.strip(), updatedate.text.strip()]
            thewriter.writerow(info)
else:
    print(f'Something went wrong for {URL} with status code : {res.status_code}.')
```
